**Task 3. Find the city having the highest number of sales**

**Extract the city**

```python
a = '359 Spruce St, Seattle, WA 98101'
```

```python
a.split(',')
```

```python
def extract_city(x):
  return x.split(',')[1][1:]
```

```python
df['City'] = df['Purchase Address'].apply(lambda x : extract_city(x))
```

```python
df
```

**Now find the highest**

```python
city_sales = df.groupby('City')['Sales'].sum()
```

```python
city_sales
```

```python
cities = df['City'].unique()
```

```python
plt.bar(cities,city_sales)
plt.xticks(cities)
plt.xticks(rotation = 90)
plt.xlabel('City')
plt.ylabel("Number")
plt.title('Chart of Sales in the US')
```

*Solution: Atlanta*

**Task 4: What time should we display advertisement to maximize likelihood of customer to by product**

**Create the hour column**

```python
def extract_hour(x):
  return x.split(" ")[1][0:2]
```

```python
df['Hour'] = df['Order Date'].apply(lambda x : extract_hour(x))
```

```python
df
```

```python
hours = df['Hour'].unique()
```

```python
hours = np.sort(hours)
```

```python
hours
```

```python
hour_sales = df.groupby('Hour')['Sales'].sum()
```

```python
plt.bar(hours,hour_sales)
plt.xticks(hours)
plt.xlabel('Hour')
plt.ylabel('USD')
```

*Solution: 18h*
